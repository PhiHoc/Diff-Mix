```python
import matplotlib.pyplot as plt
import numpy as np

# font = {'family': 'serif', 'weight': 'normal', 'size': 16}

# 示例数据
x = [0.05, 0.1, 0.3, 0.5, 0.7, 0.9, 1.0]
y = [82.54, 82.89, 82.61, 80.58, 79.22, 75.340, 63.1]
y2 = [56.24, 57.90, 59.44, 57.92, 57.16, 54.71, 36.7]

plt.figure(figsize=(8,6), dpi=80)
fig, ax1 = plt.subplots()

# 画折线图
ax1.axhline(y=81.60, linestyle='--', color='#427D9D', label='All-Baseline')
line1 = ax1.plot(x, y, marker='o', ms=10, linestyle='-', color='#427D9D', label='All')
# 添加标签和标题
ax1.set_xlabel('$p$', fontdict=font)
ax1.set_ylabel('Acc (%) - All', fontdict=font)
ax1.tick_params(axis='both', labelsize=20)
ax1.set_ylim(60, 85)

# 创建第二个 y 轴
ax2 = ax1.twinx()
line2 = ax2.plot(x, y2, marker='s', ms=10, linestyle='-', color='#A2C579', label='5shot')
ax2.axhline(y=51.60, linestyle='--', color='#A2C579', label='5shot-Baseline')
ax2.tick_params(axis='both', labelsize=20,labelcolor='#A2C579')
ax2.set_ylabel('Acc (%) - 5shot', fontdict=font)
# 显示图例
lines = line1 + line2 
labels = [line.get_label() for line in lines]
# ax1.legend(fontsize=12)
ax2.legend(lines, labels,fontsize=fontsize)
plt.xticks(fontsize=fontsize)
plt.tight_layout()
plt.savefig('../figures/ab_synp.pdf',bbox_inches='tight')
# 显示图形
plt.show()
```
